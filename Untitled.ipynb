## World Cup 2023 Data Analysis - Column Explanation

The dataset we'll be working with for the World Cup 2023 analysis contains various metrics related to both batting and bowling performances. Here's a breakdown of each column:

**Basic Information:**

* **team:** Name of the team the player belongs to.
* **player:** Name of the individual player.
* **bat_or_bowl:** Indicates whether the data pertains to batting ("bat") or bowling ("bowl").

**Performance Metrics:**

* **bb_bf (balls bowled or balls faced):** This depends on whether it's batting or bowling data. For batting, it represents the number of balls faced by the player. For bowling, it's the number of balls bowled by the player.
* **runs:** Total runs scored by the batsman or runs conceded by the bowler.
* **wkts (wickets):** Number of wickets taken by the bowler (applies only to bowling data).

**Bowling Specific Metrics:**

* **wicketball_prob:** Probability of taking a wicket with each ball bowled (might not be available in all datasets).
* **mdns (maidens):** Number of overs bowled by the bowler where no runs were conceded.
* **econ (economy rate):** Runs conceded per over bowled (runs / overs).

**Batting Specific Metrics:**

* **runs_per_ball:** Average number of runs scored per ball faced (runs / balls faced).
* **4s (fours):** Number of boundaries hit where the ball reached the boundary without bouncing.
* **6s (sixes):** Number of boundaries hit where the ball went over the boundary.
* **sr (strike rate):**  (runs scored / balls faced) * 100. Represents the batsman's scoring rate.
* **not_out:** Indicates whether the batsman remained unbeaten (1) or got dismissed (0).

**Match Information:**

* **opposition:** Name of the opposing team the player was playing against.
* **ground:** Name of the cricket ground where the match took place.
* **start_date:** Date on which the match started.
* **overs:** Total number of overs bowled in the innings the player participated in.
* **mins (minutes):** Might represent the duration of the innings or the player's batting/bowling time (depending on the data source).
* **inns (innings):** Innings number (1st or 2nd) the player participated in.

By analyzing these columns, we can gain valuable insights into team and player performances throughout the World Cup 2023. We can identify strengths and weaknesses, compare performances across different conditions and oppositions, and uncover interesting trends over the course of the tournament.

#### We will be spliting the data depending on the bat_or_bowl column

```python
bat = df[df['bat_or_bowl']=='bat']
bat.drop(['bat_or_bowl','wkts', 'overs', 'mdns', 'econ'], axis=1, inplace=True)
bat = bat.reset_index(drop=True)

display(bat.head())
```

```python
bowl = df[df['bat_or_bowl']=='bowl']
bowl.drop(['bat_or_bowl','4s','6s','sr','not_out','mins'], axis=1, inplace=True)
bowl = bowl.reset_index(drop=True)

display(bowl.head())
```

## 1- Team Performance Analysis

#### Batters & Bowlers Analysis

```python
team_data = df.groupby('team')
```

```python
bat_data = bat.groupby('team')
bowl_data = bowl.groupby('team')
```

```python
print("Ranked team scored runs:")
sns.barplot(team_data['runs'].mean().sort_values(ascending=False))
plt.show()
```

```python
print("Average runs scored by batters and bowlers:")
g = sns.FacetGrid(df, col='bat_or_bowl')
g.map(sns.barplot, 'runs', 'team')
plt.show()
```

#### Batter Analysis

```python
print('Batting average runs ranked')
bat_data['runs'].mean().sort_values(ascending=False)
```

```python
print('Batting average strike rate ranked')
bat_data['sr'].mean().sort_values(ascending=False)
```

#### Bowler Analysis

```python
print('Bowlers average runs ranked')
bowl_data['runs'].mean().sort_values(ascending=False)
```

```python
print("Total wickets taken by each team:")
bowl_data['wkts'].sum().sort_values(ascending=False)
```

```python
print('Average bowling economy rate for each team ranked')
bowl_data['econ'].sum().sort_values(ascending=False)
```

## Temporal Analysis

```python
df['start_date'] = pd.to_datetime(df['start_date'])
df['start_date']
```

```python
avg_runs_per_date = df.groupby('start_date')['runs'].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_runs_per_date.index, avg_runs_per_date.values, marker='o')

plt.xlabel('Start Date')
plt.ylabel('Average Runs per Match')
plt.title('Average Runs Scored per Match by Date (World Cup 2023)')

plt.grid(True)

plt.show()
```

```python
avg_wickets_per_date = df.groupby('start_date')['wkts'].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_wickets_per_date.index, avg_wickets_per_date.values, marker='o')

plt.xlabel('Start Date')
plt.ylabel('Average Wickets per Match')
plt.title('Average Wickets Scored per Match by Date (World Cup 2023)')

plt.grid(True)

plt.show()
```

```python
avg_runs_per_date = df.groupby(['start_date', 'team'])[['runs']].mean()

a4_dims = (11.7, 5.27)

fig, ax = plt.subplots(figsize=a4_dims)
ax.set_title('Average Team Runs Per Date')
ax.grid()

sns.lineplot(data=avg_runs_per_date, x="start_date", y="runs", hue="team", ax=ax)
```

```python
avg_wkts_per_date = df.groupby(['start_date', 'team'])[['wkts']].mean()

a4_dims = (11.7, 5.27)

fig, ax = plt.subplots(figsize=a4_dims)
ax.set_title('Average Team Wickits Per Date')
ax.grid()

sns.lineplot(data=avg_wkts_per_date, x="start_date", y="wkts", hue="team", ax=ax)
```

```python
avg_runs_per_overs = df.groupby(['overs', 'team'])[['runs']].mean()

a4_dims = (11.7, 5.27)

fig, ax = plt.subplots(figsize=a4_dims)
ax.set_title('Average Runs Per Overs')
ax.grid()

sns.lineplot(data=avg_runs_per_overs, x="overs", y="runs", hue="team", ax=ax)
```
